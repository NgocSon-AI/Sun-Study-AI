<h3>plot</h3>
"plot" is the simplest and most common plotting function, it simply creates a line plot from a list of values. (it also can take other data types and is commonly used with the numpy library - wait till the next Tutorial!)

"plot" Là hàm vẽ đồ thị đơn giản và phổ biến nhất, nó chỉ tạo một biểu đồ đường từ danh sách các giá trị (thường được sử dụng với thư viện numpy)

```python
#lets create a plot of a sine wave!
#create some empty lists
x_values = []
y_values = []

#In a for-loop create the x and y values
#use the math function "sin" to compute the sine
# Sử dụng hàm sin để tính
for i in range(100):
    x_values.append(i)
    y_values.append(math.sin(0.2*i))

plt.plot(x_values, y_values)
```

Lets add some labels and a title to the plot

Thêm nhãn và tiêu đề cho biểu đồ

```python
plt.plot(x_values, y_values)
#Y axis lable
plt.ylabel('Sine of X')
#X axis lable
plt.xlabel('X')
#Plot title
plt.title("Sine Plot")
```

We can even change some of the properties of the line

Có thể thay đổi một số thuộc tính của dòng

```python
#c = colour
#marker = line/datapoint style
plt.plot(x_values, y_values, color = "g", marker = "s")
plt.ylabel('Sine of X')
plt.xlabel('X')
plt.title("Sine Plot")
```

<h3>scatter</h3>
Another common type of plot used is the scatter plot!

Một loại biểu đồ phổ biến khác là biểu đồ scatter

```python
#lets create a plot of a noisy parabola!
#create some empty lists
x_parab_values = []
y_parab_values = []

#In a for-loop create the x and y values
#use the math function "pow" to compute x^2
#and then add a random number from 0 - 1
for i in range(-50, 50):
    x_parab_values.append(i/20)
    y_parab_values.append(math.pow(i/20, 2) + rd.random())

# Vẽ parabol
```

```python
plt.scatter(x_parab_values, y_parab_values, c = "k", marker = "*")
plt.ylabel('Y')
plt.xlabel('X')
plt.title("Noisy Parabola")
```

```python
import numpy as np
```

```python
t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 4 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Vẽ hình trái tim
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red', marker=".")
plt.fill(x, y, color='red', alpha=0.5)
plt.title('Heart Shape')
plt.axis('equal')  # Đảm bảo tỷ lệ x và y giống nhau
plt.grid(False)
plt.show()
```

<h3>Figures</h3>
These plots are a bit small, Lets increase the size!

Phóng to 

```python
plt.figure(figsize=(20, 10))
plt.scatter(x_parab_values, y_parab_values, c = "k", marker = "*")
plt.ylabel('Y')
plt.xlabel('X')
plt.title("Noisy Parabola")
```

 "plt.figure" changes the properties of the "current" figure. In the case of using Jupyter notebooks this is whatever we are plotting in the current cell (it won't overwrite other cells that have already plotted) but if we only have one figure matplotlib will plot everything on the same figure.

 Thay đổi thuộc tính của hình hiện tại. Nếu chỉ có một hình matplotlib sẽ vẽ mọi thứ trên cùng một hình.

```python
plt.scatter(x_parab_values, y_parab_values, c = "k", marker = "*")
plt.ylabel('Y')
plt.xlabel('X')
plt.title("Noisy Parabola")

plt.plot(x_values, y_values, c = "r", marker = "o")
plt.ylabel('Sine of X')
plt.xlabel('X')
plt.title("Sine Plot")
```

<h3>Creating multiple figures</h3>
One way we can easily get around this mess is by creating multiple figures

Tạo nhiều hình ảnh.
Một cách chúng ta có thể giải quyết vấn đề trên là tạo nhiều hình ảnh.

```python
#Create a figure
plt.figure(1, figsize=(20, 10))
plt.scatter(x_parab_values, y_parab_values, c = "k", marker = "*")
plt.ylabel('Y')
plt.xlabel('X')
plt.title("Noisy Parabola")

#Create a new figure
plt.figure(2, figsize=(20, 10))
plt.plot(x_values, y_values, c = "r", marker = "o")
plt.ylabel('Sine of X')
plt.xlabel('X')
plt.title("Sine Plot")
```

<h3>Creating Subplots</h3>
Now let's look at is how we can plot multiple graphs independantly on the same figure

Vẽ nhiều đồ thị độc lập trên cùng một hình.

```python
plt.figure(1, figsize=(10, 10))
plt.subplot(211)
plt.scatter(x_parab_values, y_parab_values, c = "k", marker = "*")
plt.ylabel('Y')
plt.xlabel('X')
plt.title("Noisy Parabola")

plt.subplot(212)
plt.plot(x_values, y_values, c = "r", marker = "o")
plt.ylabel('Sine of X')
plt.xlabel('X')
plt.title("Sine Plot")
```
